```python
from gensim.models import Word2Vec
sentences = [["cat", "say", "meow"], ["dog", "say", "woof"]]
model = Word2Vec(sentences, min_count=1)
```

```python
model.wv.vocab
```

```python
vector = model.wv['dog']  # numpy vector of a word
vector
```

```python
model.most_similar('cat')
```

```python
from gensim.models.word2vec import LineSentence
import pandas as pd
```

```python
database = 'D:/Machine Learning/NLP/data/sqlResult_1558435.csv'
content = pd.read_csv(database, encoding='gb18030')
sample = content['content'][:100].tolist()
```

```python
import jieba
def cut(string):
    return ' '.join(jieba.cut(string))
```

```python
with open('mini_samples.txt', 'w') as f:
    for s in sample:
        f.write(cut(s) + '\n')
```
